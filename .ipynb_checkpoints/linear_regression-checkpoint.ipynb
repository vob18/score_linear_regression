# 2. Preparar los datos

```python
# tomar todos los valores menos student_id
x_aux =  data[data.columns.difference(['student_id','posttest'], sort=False)]

# Agregar el término independiente
x_aux = sm.add_constant(x_aux)
```

```python
y = data.posttest
```

#### X para la prueba de control

```python
Xa = pd.get_dummies(data = x_aux, drop_first=True)
```

#### X para la prueba que NO toma la variable *classroom*

```python
Xc = x_aux[x_aux.columns.difference(['classroom'], sort=False)]
Xc = pd.get_dummies(data = Xc, drop_first=True)
```

#### X para la prueba que NO toma la variable *school*

```python
Xs = x_aux[x_aux.columns.difference(['school'], sort=False)]
Xs = pd.get_dummies(data = Xs, drop_first=True)
```

# 3. Validación Cruzada

```python
kf = KFold(n_splits = 10, shuffle = True, random_state=50)
```

### 3.1 Prueba con todas las variables predictoras

```python
a_scores = []
a_preds =[]

a_mse = []
```

```python
%%timeit -r 5 -n 100
for train_index, test_index in kf.split(Xa):
    
    X_train, X_test, y_train, y_test = Xa.loc[train_index],Xa.loc[test_index],y.loc[train_index],y.loc[test_index]
    
    a_model = sm.OLS(y_train, X_train).fit()
    
    # métricas del modelo
    a_scores.append(a_model.rsquared_adj)
    pred = a_model.predict(X_test)
    a_preds.append(pred)
    a_mse.append(metrics.mean_squared_error(y_test,pred))
```

### 3.2 Prueba eliminando la variable *classroom*

```python
c_scores = []
c_preds = []

c_mse = []
```

```python
%%timeit -r 5 -n 100
for train_index, test_index in kf.split(Xc):
    
    X_train, X_test, y_train, y_test = Xc.loc[train_index],Xc.loc[test_index],y.loc[train_index],y.loc[test_index]
    
    c_model = sm.OLS(y_train, X_train).fit()
    
    # métricas del modelo
    c_scores.append(c_model.rsquared_adj)
    pred = c_model.predict(X_test)
    c_preds.append(pred)
    c_mse.append(metrics.mean_squared_error(y_test,pred))
```

### 3.3 Prueba eliminando la variable *school*

```python
s_scores = []
s_preds = []

s_mse = []
```

```python
%%timeit -r 5 -n 100
for train_index, test_index in kf.split(Xs):
    
    X_train, X_test, y_train, y_test = Xs.loc[train_index],Xs.loc[test_index],y.loc[train_index],y.loc[test_index]
    
    s_model = sm.OLS(y_train, X_train).fit()
       
    # métricas del modelo
    s_scores.append(s_model.rsquared_adj)
    pred = s_model.predict(X_test)
    s_preds.append(pred)
    s_mse.append(metrics.mean_squared_error(y_test,pred))
```

### 3.4 Puntajes

```python
scores = [['all',np.mean(a_scores),np.mean(a_mse)],
          ['without_classroom',np.mean(c_scores),np.mean(c_mse)],
          ['without_school',np.mean(s_scores),np.mean(s_mse)]]
pd.DataFrame(data = scores, columns=['Combination','R^2 adjusted','MSE'])
```

### 3.5 Predicciones

Debido al cell magic %%timeit, el modelo **s_model**, generado dentro de la celda no es accesible pero gracias a que el valor de random state de la funcion KFold fue establecida en 50 se puede repetir la validación y obtner los mismos resultados 

```python
for train_index, test_index in kf.split(Xs):
    
    X_train, X_test, y_train, y_test = Xs.loc[train_index],Xs.loc[test_index],y.loc[train_index],y.loc[test_index]
    
    s_model = sm.OLS(y_train, X_train).fit()
```

```python
# predicciones con el última iteración (fold) de la validación
pred = s_model.predict(X_test)
pd.DataFrame(data = {'Target':y_test, 'Prediction':pred})
```

verificar que los resultados de la predcicción anterior coinciden con el último arreglo de las predcciónes guardado en **s_preds**
